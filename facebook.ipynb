```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from math import sqrt
from sklearn.metrics import mean_squared_error

# For LSTM / Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# For Prophet
from prophet import Prophet

def train_test_split_time(
    df, 
    date_col='InvoiceDate', 
    test_ratio=0.2
):
    """
    Split the time series data by date in ratio test_ratio.
    We assume df is sorted by date already.
    """
    df = df.sort_values(by=date_col).reset_index(drop=True)
    nrows = len(df)
    cutoff = int((1 - test_ratio) * nrows)

    df_train = df.iloc[:cutoff].copy()
    df_test = df.iloc[cutoff:].copy()
    return df_train, df_test

def create_X_y_lstm(df, feature_cols, target_col='sales', sequence_length=3):
    """
    Prepare sequences for LSTM input.
    feature_cols is a list of columns used as input features.
    target_col is the column we want to predict.
    sequence_length indicates how many time steps in each input sequence.

    Returns: 
        X (numpy array): shape (num_samples, sequence_length, num_features)
        y (numpy array): shape (num_samples,)
    """
    data = df[feature_cols + [target_col]].copy()
    values = data.values
    X, y = [], []
    for i in range(len(values) - sequence_length):
        # Slices [i : i+sequence_length] for features
        X_seq = values[i : i + sequence_length, : len(feature_cols)]
        # The target is the next time step's 'target_col'
        y_val = values[i + sequence_length, -1] 
        X.append(X_seq)
        y.append(y_val)
    X = np.array(X)
    y = np.array(y)
    return X, y

def build_lstm_model(num_layers=1, units=32, dropout_rate=0.0, activation='relu', input_shape=(3,5)):
    """
    Build a simple LSTM model with the specified hyperparameters.

    num_layers: number of LSTM layers
    units: number of LSTM units (neurons) per layer
    dropout_rate: dropout ratio for the LSTM
    activation: activation function
    input_shape: (sequence_length, num_features)
    """
    model = Sequential()
    for layer_idx in range(num_layers):
        is_last = (layer_idx == num_layers - 1)
        return_seqs = not is_last
        lstm_kwargs = {
            'units': units,
            'activation': activation,
            'return_sequences': return_seqs,
        }
        if layer_idx == 0:
            lstm_kwargs['input_shape'] = input_shape

        model.add(LSTM(**lstm_kwargs))
        model.add(Dropout(dropout_rate))

    model.add(Dense(1))  # final output layer for regression
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

def _compute_error_metrics(y_true, y_pred):
    """
    Compute RMSE, MAE, and MAPE for given true and predicted values.
    Returns a dict with keys: 'rmse', 'mae', 'mape'.
    """
    rmse = sqrt(mean_squared_error(y_true, y_pred))
    mae = np.mean(np.abs(y_true - y_pred))
    # To avoid division by zero, filter out zero or near-zero actuals
    nonzero_mask = (y_true != 0)
    if np.any(nonzero_mask):
        mape = np.mean(np.abs((y_true[nonzero_mask] - y_pred[nonzero_mask]) 
                              / y_true[nonzero_mask])) * 100
    else:
        mape = np.nan
    return {'rmse': rmse, 'mae': mae, 'mape': mape}

def evaluate_lstm_model(df_train, df_test, date_col='InvoiceDate', sales_col='sales'):
    """
    Perform a small hyperparameter search for LSTM, returning:
      - df_test_pred with predictions,
      - dictionary with best_config, valRMSE/testRMSE, valMAE/testMAE, valMAPE/testMAPE
    We'll do an internal 'val' split from df_train (80/20).
    """
    # 1) Create an internal train/val split from df_train
    df_tr, df_val = train_test_split_time(df_train, date_col=date_col, test_ratio=0.2)

    # 2) Choose features for LSTM
    feature_cols = ['sales_lag1','sales_lag2','sales_lag3','rolling_mean_3','rolling_mean_6']

    # Drop NaNs in these feature_cols + sales_col
    df_tr = df_tr.dropna(subset=feature_cols + [sales_col])
    df_val = df_val.dropna(subset=feature_cols + [sales_col])
    df_test = df_test.dropna(subset=feature_cols + [sales_col])

    # We'll set sequence_length=1 because we already have manual lag features
    sequence_length = 1

    # 3) Create X,y for train, val, test
    X_tr, y_tr = create_X_y_lstm(df_tr, feature_cols, sales_col, sequence_length=sequence_length)
    X_val, y_val = create_X_y_lstm(df_val, feature_cols, sales_col, sequence_length=sequence_length)
    X_test, y_test = create_X_y_lstm(df_test, feature_cols, sales_col, sequence_length=sequence_length)

    # 4) Hyperparameter search
    possible_num_layers = [1, 2]
    possible_units = [16, 32]
    possible_dropout = [0.0, 0.2]
    possible_activations = ['relu','tanh']

    best_config = None
    best_rmse_val = 1e15
    best_model = None
    best_metrics_val = None

    for nl in possible_num_layers:
        for un in possible_units:
            for dr in possible_dropout:
                for act in possible_activations:
                    model = build_lstm_model(
                        num_layers=nl, 
                        units=un, 
                        dropout_rate=dr, 
                        activation=act,
                        input_shape=(sequence_length, len(feature_cols))
                    )
                    model.fit(X_tr, y_tr, epochs=50, batch_size=8, verbose=0)

                    # Evaluate on val
                    val_pred = model.predict(X_val)
                    metrics_val = _compute_error_metrics(y_val, val_pred)

                    if metrics_val['rmse'] < best_rmse_val:
                        best_rmse_val = metrics_val['rmse']
                        best_config = {'num_layers': nl, 'units': un, 
                                       'dropout_rate': dr, 'activation': act}
                        best_model = model
                        best_metrics_val = metrics_val

    # Evaluate on test
    test_pred = best_model.predict(X_test)
    metrics_test = _compute_error_metrics(y_test, test_pred)

    # Create df_test_pred
    offset_test = sequence_length
    df_test_pred = df_test.iloc[offset_test:].copy()
    df_test_pred['LSTM_Pred'] = test_pred.flatten()

    info = {
        'best_config': best_config,
        'valRMSE': best_metrics_val['rmse'],
        'valMAE': best_metrics_val['mae'],
        'valMAPE': best_metrics_val['mape'],
        'testRMSE': metrics_test['rmse'],
        'testMAE': metrics_test['mae'],
        'testMAPE': metrics_test['mape']
    }
    return df_test_pred, info

def evaluate_prophet_model(df_train, df_test, date_col='InvoiceDate', sales_col='sales'):
    """
    Perform a small hyperparameter search for Prophet,
    returning:
      - df_test_pred with 'Prophet_Pred',
      - a dict with best_config + (valRMSE/testRMSE, valMAE/testMAE, valMAPE/testMAPE).
    We'll do an internal train/val split from df_train (80/20).
    """
    # Internal split
    df_tr, df_val = train_test_split_time(df_train, date_col=date_col, test_ratio=0.2)

    # Drop NaN
    df_tr = df_tr.dropna(subset=[sales_col])
    df_val = df_val.dropna(subset=[sales_col])
    df_test = df_test.dropna(subset=[sales_col])

    # Convert to Prophet format: ds, y
    def to_prophet(df):
        return df.rename(columns={date_col: 'ds', sales_col: 'y'})[['ds','y']]
    prophet_tr = to_prophet(df_tr)
    prophet_val = to_prophet(df_val)
    prophet_test = to_prophet(df_test)

    possible_seasonality_modes = ['additive', 'multiplicative']
    possible_seasonality_prior_scale = [1.0, 5.0]

    best_config = None
    best_rmse_val = 1e15
    best_model = None
    best_metrics_val = None

    for mode in possible_seasonality_modes:
        for prior_scale in possible_seasonality_prior_scale:
            m = Prophet(
                seasonality_mode=mode,
                seasonality_prior_scale=prior_scale,
                yearly_seasonality=True,
                weekly_seasonality=False,
                daily_seasonality=False
            )
            m.fit(prophet_tr)

            # Predict on val
            future_val = prophet_val[['ds']].copy()
            val_forecast = m.predict(future_val)
            val_pred = val_forecast['yhat'].values
            y_val = prophet_val['y'].values

            metrics_val = _compute_error_metrics(y_val, val_pred)
            if metrics_val['rmse'] < best_rmse_val:
                best_rmse_val = metrics_val['rmse']
                best_config = {
                    'seasonality_mode': mode,
                    'seasonality_prior_scale': prior_scale
                }
                best_model = m
                best_metrics_val = metrics_val

    if best_model is None:
        # Means no data or some other issue
        return df_test.copy(), {
            'best_config': None,
            'valRMSE': float('inf'),
            'valMAE': None,
            'valMAPE': None,
            'testRMSE': None,
            'testMAE': None,
            'testMAPE': None
        }

    # Evaluate on test
    future_test = prophet_test[['ds']].copy()
    test_forecast = best_model.predict(future_test)
    test_pred = test_forecast['yhat'].values
    y_test = prophet_test['y'].values
    metrics_test = _compute_error_metrics(y_test, test_pred)

    df_test_pred = df_test.copy()
    df_test_pred['Prophet_Pred'] = test_pred

    info = {
        'best_config': best_config,
        'valRMSE': best_metrics_val['rmse'],
        'valMAE': best_metrics_val['mae'],
        'valMAPE': best_metrics_val['mape'],
        'testRMSE': metrics_test['rmse'],
        'testMAE': metrics_test['mae'],
        'testMAPE': metrics_test['mape']
    }
    return df_test_pred, info

def main():
    folder_path = "/Users/ton.w/Documents/Columbia/forecast/group project/individual_group"
    pattern = "*_monthly_sales_with_lags.csv"
    csv_files = glob.glob(os.path.join(folder_path, pattern))

    # Also add the single combined file if needed
    special_file = os.path.join(folder_path, "monthly_sales_after_filter_with_lags.csv")
    if os.path.exists(special_file):
        csv_files.append(special_file)

    for csv_file in csv_files:
        print(f"\n=== Processing file: {os.path.basename(csv_file)} ===")
        df = pd.read_csv(csv_file)

        # Ensure date col is datetime
        if 'InvoiceDate' in df.columns:
            df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

        # Drop rows with NaN in 'sales'
        if 'sales' in df.columns:
            df = df.dropna(subset=['sales'])
        else:
            print("No 'sales' column found, skipping.")
            continue

        # Train/test split
        df_train, df_test = train_test_split_time(df, date_col='InvoiceDate', test_ratio=0.2)

        # 1) LSTM
        df_test_pred_lstm, lstm_info = evaluate_lstm_model(
            df_train, df_test, date_col='InvoiceDate', sales_col='sales'
        )
        print("LSTM best config:", lstm_info['best_config'])
        print(f"LSTM validation metrics -> RMSE={lstm_info['valRMSE']:.2f}, MAE={lstm_info['valMAE']:.2f}, MAPE={lstm_info['valMAPE']:.2f}")
        print(f"LSTM test metrics       -> RMSE={lstm_info['testRMSE']:.2f}, MAE={lstm_info['testMAE']:.2f}, MAPE={lstm_info['testMAPE']:.2f}")

        # 2) Prophet
        df_test_pred_prophet, prophet_info = evaluate_prophet_model(
            df_train, df_test, date_col='InvoiceDate', sales_col='sales'
        )
        print("Prophet best params:", prophet_info['best_config'])
        print(f"Prophet validation metrics -> RMSE={prophet_info['valRMSE']:.2f}, MAE={prophet_info['valMAE']:.2f}, MAPE={prophet_info['valMAPE']:.2f}")
        print(f"Prophet test metrics       -> RMSE={prophet_info['testRMSE']:.2f}, MAE={prophet_info['testMAE']:.2f}, MAPE={prophet_info['testMAPE']:.2f}")

        # Merge predictions
        df_test_pred = pd.merge(
            df_test_pred_lstm[['InvoiceDate','sales','LSTM_Pred']],
            df_test_pred_prophet[['InvoiceDate','Prophet_Pred']],
            on='InvoiceDate', how='outer'
        )

        # Plot
        fig, ax = plt.subplots(figsize=(8,5))
        plt.plot(df['InvoiceDate'], df['sales'], 'o-', label='Actual')

        first_test_date = df_test['InvoiceDate'].iloc[0]
        plt.axvline(x=first_test_date, color='gray', linestyle='--', label='Train/Test Split')

        # LSTM predictions
        if 'LSTM_Pred' in df_test_pred.columns:
            plt.plot(df_test_pred['InvoiceDate'], df_test_pred['LSTM_Pred'], 'x--', 
                     label=f"LSTM (RMSE={lstm_info['testRMSE']:.0f})")

        # Prophet predictions
        if 'Prophet_Pred' in df_test_pred.columns:
            plt.plot(df_test_pred['InvoiceDate'], df_test_pred['Prophet_Pred'], 's--', 
                     label=f"Prophet (RMSE={prophet_info['testRMSE']:.0f})")

        plt.title(f"{os.path.basename(csv_file)}: LSTM vs Prophet (tuned)")
        plt.xlabel("Date")
        plt.ylabel("Sales")
        plt.legend()
        plt.show()

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from math import sqrt
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

from prophet import Prophet

# ---------------------
# 1) Helper functions
# ---------------------

def forecast_bias(y_true, y_pred):
    """
    Computes forecast bias as:
      sum(Actual - Forecast) / sum(Forecast) * 100
    """
    return (np.sum(y_true - y_pred) / np.sum(y_pred)) * 100

def train_test_split_time(df, date_col='InvoiceDate', test_ratio=0.2):
    df = df.sort_values(by=date_col).reset_index(drop=True)
    nrows = len(df)
    cutoff = int((1 - test_ratio) * nrows)
    df_train = df.iloc[:cutoff].copy()
    df_test = df.iloc[cutoff:].copy()
    return df_train, df_test

def create_X_y_lstm(df, feature_cols, target_col='sales', sequence_length=3):
    data = df[feature_cols + [target_col]].copy()
    values = data.values
    X, y = [], []
    for i in range(len(values) - sequence_length):
        X_seq = values[i : i + sequence_length, : len(feature_cols)]
        y_val = values[i + sequence_length, -1]
        X.append(X_seq)
        y.append(y_val)
    return np.array(X), np.array(y)

def build_lstm_model(num_layers=1, units=32, dropout_rate=0.0, activation='relu', input_shape=(3,5)):
    model = Sequential()
    for layer_idx in range(num_layers):
        is_last = (layer_idx == num_layers - 1)
        return_seqs = not is_last
        lstm_kwargs = {
            'units': units,
            'activation': activation,
            'return_sequences': return_seqs,
        }
        if layer_idx == 0:
            lstm_kwargs['input_shape'] = input_shape

        model.add(LSTM(**lstm_kwargs))
        model.add(Dropout(dropout_rate))

    model.add(Dense(1))  # final output layer
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

def _compute_error_metrics(y_true, y_pred):
    rmse = sqrt(mean_squared_error(y_true, y_pred))
    mae = np.mean(np.abs(y_true - y_pred))
    nonzero_mask = (y_true != 0)
    if np.any(nonzero_mask):
        mape = np.mean(np.abs((y_true[nonzero_mask] - y_pred[nonzero_mask]) / y_true[nonzero_mask])) * 100
    else:
        mape = np.nan
    return {'rmse': rmse, 'mae': mae, 'mape': mape}

def evaluate_lstm_model(df_train, df_test, date_col='InvoiceDate', sales_col='sales'):
    # Internal train/val split
    df_tr, df_val = train_test_split_time(df_train, date_col=date_col, test_ratio=0.2)
    feature_cols = ['sales_lag1','sales_lag2','sales_lag3','rolling_mean_3','rolling_mean_6']

    df_tr = df_tr.dropna(subset=feature_cols + [sales_col])
    df_val = df_val.dropna(subset=feature_cols + [sales_col])
    df_test = df_test.dropna(subset=feature_cols + [sales_col])

    sequence_length = 1
    X_tr, y_tr = create_X_y_lstm(df_tr, feature_cols, sales_col, sequence_length)
    X_val, y_val = create_X_y_lstm(df_val, feature_cols, sales_col, sequence_length)
    X_test, y_test = create_X_y_lstm(df_test, feature_cols, sales_col, sequence_length)

    possible_num_layers = [1, 2]
    possible_units = [16, 32]
    possible_dropout = [0.0, 0.2]
    possible_activations = ['relu','tanh']

    best_config = None
    best_rmse_val = 1e15
    best_model = None
    best_metrics_val = None

    for nl in possible_num_layers:
        for un in possible_units:
            for dr in possible_dropout:
                for act in possible_activations:
                    model = build_lstm_model(
                        num_layers=nl, 
                        units=un, 
                        dropout_rate=dr, 
                        activation=act,
                        input_shape=(sequence_length, len(feature_cols))
                    )
                    model.fit(X_tr, y_tr, epochs=50, batch_size=8, verbose=0)

                    val_pred = model.predict(X_val)
                    metrics_val = _compute_error_metrics(y_val, val_pred)

                    if metrics_val['rmse'] < best_rmse_val:
                        best_rmse_val = metrics_val['rmse']
                        best_config = {'num_layers': nl, 'units': un, 
                                       'dropout_rate': dr, 'activation': act}
                        best_model = model
                        best_metrics_val = metrics_val

    test_pred = best_model.predict(X_test)
    metrics_test = _compute_error_metrics(y_test, test_pred)

    offset_test = sequence_length
    df_test_pred = df_test.iloc[offset_test:].copy()
    df_test_pred['LSTM_Pred'] = test_pred.flatten()

    info = {
        'best_config': best_config,
        'valRMSE': best_metrics_val['rmse'],
        'valMAE': best_metrics_val['mae'],
        'valMAPE': best_metrics_val['mape'],
        'testRMSE': metrics_test['rmse'],
        'testMAE': metrics_test['mae'],
        'testMAPE': metrics_test['mape']
    }
    return df_test_pred, info

def evaluate_prophet_model(df_train, df_test, date_col='InvoiceDate', sales_col='sales'):
    df_tr, df_val = train_test_split_time(df_train, date_col=date_col, test_ratio=0.2)

    df_tr = df_tr.dropna(subset=[sales_col])
    df_val = df_val.dropna(subset=[sales_col])
    df_test = df_test.dropna(subset=[sales_col])

    def to_prophet(df_):
        return df_.rename(columns={date_col: 'ds', sales_col: 'y'})[['ds','y']]
    prophet_tr = to_prophet(df_tr)
    prophet_val = to_prophet(df_val)
    prophet_test = to_prophet(df_test)

    possible_seasonality_modes = ['additive', 'multiplicative']
    possible_seasonality_prior_scale = [1.0, 5.0]

    best_config = None
    best_rmse_val = 1e15
    best_model = None
    best_metrics_val = None

    for mode in possible_seasonality_modes:
        for prior_scale in possible_seasonality_prior_scale:
            m = Prophet(
                seasonality_mode=mode,
                seasonality_prior_scale=prior_scale,
                yearly_seasonality=True,
                weekly_seasonality=False,
                daily_seasonality=False
            )
            m.fit(prophet_tr)

            future_val = prophet_val[['ds']].copy()
            val_forecast = m.predict(future_val)
            val_pred = val_forecast['yhat'].values
            y_val = prophet_val['y'].values

            metrics_val = _compute_error_metrics(y_val, val_pred)
            if metrics_val['rmse'] < best_rmse_val:
                best_rmse_val = metrics_val['rmse']
                best_config = {'seasonality_mode': mode, 'seasonality_prior_scale': prior_scale}
                best_model = m
                best_metrics_val = metrics_val

    if best_model is None:
        return df_test.copy(), {
            'best_config': None,
            'valRMSE': float('inf'),
            'valMAE': None,
            'valMAPE': None,
            'testRMSE': None,
            'testMAE': None,
            'testMAPE': None
        }

    future_test = prophet_test[['ds']].copy()
    test_forecast = best_model.predict(future_test)
    test_pred = test_forecast['yhat'].values
    y_test = prophet_test['y'].values
    metrics_test = _compute_error_metrics(y_test, test_pred)

    df_test_pred = df_test.copy()
    df_test_pred['Prophet_Pred'] = test_pred

    info = {
        'best_config': best_config,
        'valRMSE': best_metrics_val['rmse'],
        'valMAE': best_metrics_val['mae'],
        'valMAPE': best_metrics_val['mape'],
        'testRMSE': metrics_test['rmse'],
        'testMAE': metrics_test['mae'],
        'testMAPE': metrics_test['mape']
    }
    return df_test_pred, info


# ---------------------
# 2) MAIN LOGIC FOR A SINGLE FILE
#    (just the total data file)
# ---------------------

# Path to the single combined dataset
csv_file = "/Users/ton.w/Documents/Columbia/forecast/group project/individual_group/monthly_sales_after_filter_with_lags.csv"

df = pd.read_csv(csv_file)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df = df.dropna(subset=['sales'])

# Split train/test
df_train, df_test = train_test_split_time(df, date_col='InvoiceDate', test_ratio=0.2)

# Evaluate LSTM
df_test_pred_lstm, lstm_info = evaluate_lstm_model(
    df_train, df_test, date_col='InvoiceDate', sales_col='sales'
)
print("=== LSTM Results ===")
print("Best config:", lstm_info['best_config'])
print(f"Validation metrics -> RMSE={lstm_info['valRMSE']:.2f}, MAE={lstm_info['valMAE']:.2f}, MAPE={lstm_info['valMAPE']:.2f}")
print(f"Test metrics       -> RMSE={lstm_info['testRMSE']:.2f}, MAE={lstm_info['testMAE']:.2f}, MAPE={lstm_info['testMAPE']:.2f}")

# Evaluate Prophet
df_test_pred_prophet, prophet_info = evaluate_prophet_model(
    df_train, df_test, date_col='InvoiceDate', sales_col='sales'
)
print("\n=== Prophet Results ===")
print("Best params:", prophet_info['best_config'])
print(f"Validation metrics -> RMSE={prophet_info['valRMSE']:.2f}, MAE={prophet_info['valMAE']:.2f}, MAPE={prophet_info['valMAPE']:.2f}")
print(f"Test metrics       -> RMSE={prophet_info['testRMSE']:.2f}, MAE={prophet_info['testMAE']:.2f}, MAPE={prophet_info['testMAPE']:.2f}")

# Merge test predictions for final test portion
df_test_pred = pd.merge(
    df_test_pred_lstm[['InvoiceDate','sales','LSTM_Pred']],
    df_test_pred_prophet[['InvoiceDate','Prophet_Pred']],
    on='InvoiceDate', how='outer'
)

# (Optional) Plot
fig, ax = plt.subplots(figsize=(8,5))
plt.plot(df['InvoiceDate'], df['sales'], 'o-', label='Actual')

first_test_date = df_test['InvoiceDate'].iloc[0]
plt.axvline(x=first_test_date, color='gray', linestyle='--', label='Train/Test Split')

plt.plot(df_test_pred['InvoiceDate'], df_test_pred['LSTM_Pred'], 'x--', label=f"LSTM (RMSE={lstm_info['testRMSE']:.0f})")
plt.plot(df_test_pred['InvoiceDate'], df_test_pred['Prophet_Pred'], 's--', label=f"Prophet (RMSE={prophet_info['testRMSE']:.0f})")

plt.title("Total Items: LSTM vs Prophet (tuned)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()

# ---------------------
# 3) COMPUTE FORECAST BIAS
# ---------------------
df_test_pred = df_test_pred.dropna(subset=['sales','LSTM_Pred','Prophet_Pred'])

y_true = df_test_pred['sales'].values
y_lstm = df_test_pred['LSTM_Pred'].values
y_prophet = df_test_pred['Prophet_Pred'].values

def forecast_bias(y_true, y_pred):
    return (np.sum(y_true - y_pred) / np.sum(y_pred)) * 100

bias_lstm = forecast_bias(y_true, y_lstm)
bias_prophet = forecast_bias(y_true, y_prophet)

print(f"\n=== Forecast Bias ===")
print(f"LSTM Forecast Bias: {bias_lstm:.2f}%")
print(f"Prophet Forecast Bias: {bias_prophet:.2f}%")
```
